```python
import sympy as sp
from sympy import pi
from sympy import Matrix

#Definir las bases (e1,e2,e3)

e1 = sp.Matrix([1,0,0])
e2 = sp.Matrix([0,1,0])
e3 = sp.Matrix([0,0,1])

#Definir los vectores del Punto
a = sp.Matrix([1,2,3])
b = sp.Matrix([4,5,6])
c = sp.Matrix([3,2,1])
d = sp.Matrix([6,5,4])

#Definir las normas de los vectores del punto

norm_a = a.norm()
norm_b = b.norm()
norm_c = c.norm()
norm_d = d.norm()

#Definir las normas de los vectores base

norm_e1 = e1.norm()
norm_e2 = e2.norm()
norm_e3 = e3.norm()
```

```python
#Punto (a)

#1) a + b + c d

print("a + b + c + d =", a + b + c + d)

#2) a + b - c -d

print("a + b - c - d =", a + b - c - d)

#3) a - b + c -d

print("a - b + c - d =", a - b + c - d)

#4) -a + b -c +d

print("-a + b - c + d =", -a + b - c + d)
```

```python
#Punto (b)

#1) angulo entre a y e1

dot_productae1=a.dot(e1)
norm_a=a.norm()
norm_e1=e1.norm()

angulo_ae1=sp.acos(dot_productae1/(norm_a*norm_e1))
print("Angulo entre a y e1:", angulo_ae1)

print("Ángulo entre a y e1 en grados:", (angulo_ae1 * 180 / pi).evalf())

#2) angulo entre a y e2

dot_productae2=a.dot(e2)
norm_a=a.norm()
norm_e2=e2.norm()

angulo_ae2=sp.acos(dot_productae2/(norm_a*norm_e2))
print("Angulo entre a y e2:", angulo_ae2)
print("Ángulo entre a y e2 en grados:", (angulo_ae2 * 180 / pi).evalf())

#3) angulo entre a y e3
dot_productae3=a.dot(e3)
norm_a=a.norm()
norm_e3=e3.norm()

angulo_ae3=sp.acos(dot_productae3/(norm_a*norm_e3))
print("Angulo entre a y e3:", angulo_ae3)
print("Ángulo entre a y e3 en grados:", (angulo_ae3 * 180 / pi).evalf())

#4) angulo entre b y e1
dot_productbe1=b.dot(e1)
norm_b=b.norm()
norm_e1=e1.norm()

angulo_be1=sp.acos(dot_productbe1/(norm_b*norm_e1))
print("Angulo entre b y e1:", angulo_be1)
print("Ángulo entre b y e1 en grados:", (angulo_be1 * 180 / pi).evalf())

#5) angulo entre b y e2
dot_productbe2=b.dot(e2)
norm_b=b.norm()
norm_e2=e2.norm()

angulo_be2=sp.acos(dot_productbe2/(norm_b*norm_e2))
print("Angulo entre b y e2:", angulo_be2)
print("Ángulo entre b y e2 en grados:", (angulo_be2 * 180 / pi).evalf())

#6) angulo entre b y e3
dot_productbe3=b.dot(e3)
norm_b=b.norm()
norm_e3=e3.norm()

angulo_be3=sp.acos(dot_productbe3/(norm_b*norm_e3))
print("Angulo entre b y e3:", angulo_be3)
print("Ángulo entre b y e3 en grados:", (angulo_be3 * 180 / pi).evalf())

#7) angulo entre c y e1
dot_productce1=c.dot(e1)
norm_c=c.norm()
norm_e1=e1.norm()

angulo_ce1=sp.acos(dot_productce1/(norm_c*norm_e1))
print("Angulo entre c y e1:", angulo_ce1)
print("Ángulo entre c y e1 en grados:", (angulo_ce1 * 180 / pi).evalf())

#8) angulo entre c y e2
dot_productce2=c.dot(e2)
norm_c=c.norm()
norm_e2=e2.norm()

angulo_ce2=sp.acos(dot_productce2/(norm_c*norm_e2))
print("Angulo entre c y e2:", angulo_ce2)
print("Ángulo entre c y e2 en grados:", (angulo_ce2 * 180 / pi).evalf())

#9) angulo entre c y e3
dot_productce3=c.dot(e3)
norm_c=c.norm()
norm_e3=e3.norm()

angulo_ce3=sp.acos(dot_productce3/(norm_c*norm_e3))
print("Angulo entre c y e3:", angulo_ce3)
print("Ángulo entre c y e3 en grados:", (angulo_ce3 * 180 / pi).evalf())

#10) angulo entre d y e1
dot_productde1=d.dot(e1)
norm_d=d.norm()
norm_e1=e1.norm()

angulo_de1=sp.acos(dot_productde1/(norm_d*norm_e1))
print("Angulo entre d y e1:", angulo_de1)
print("Ángulo entre d y e1 en grados:", (angulo_de1 * 180 / pi).evalf())

#11) angulo entre d y e2
dot_productde2=d.dot(e2)
norm_d=d.norm()
norm_e2=e2.norm()

angulo_de2=sp.acos(dot_productde2/(norm_d*norm_e2))
print("Angulo entre d y e2:", angulo_de2)
print("Ángulo entre d y e2 en grados:", (angulo_de2 * 180 / pi).evalf())

#12) angulo entre d y e3
dot_productde3=d.dot(e3)
norm_d=d.norm()
norm_e3=e3.norm()

angulo_de3=sp.acos(dot_productde3/(norm_d*norm_e3))
print("Angulo entre d y e3:", angulo_de3)
print("Ángulo entre d y e3 en grados:", (angulo_de3 * 180 / pi).evalf())
```

```python
#punto (h)

#Productos cruz

#1) a x b

F = a.cross(b)
print("El producto cruz de a y b es:", F)

#2) b x c

G = b.cross(c)
print("El producto cruz de b y c es:", G)

#3) c x d

H = c.cross(d)
print("El cruz de c y d es:", H)

#Angulo con d

#1) Angulo entre F y d

norm_F=F.norm()

angulo_Fd=sp.acos(F.dot(d)/(norm_F*norm_d))
print("Angulo entre F y d:", angulo_Fd)
print("Ángulo entre F y d en grados:", (angulo_Fd * 180 / pi).evalf())

#2) Angulo entre G y d

norm_G=G.norm()

angulo_Gd=sp.acos(G.dot(d)/(norm_G*norm_d))
print("Angulo entre G y d:", angulo_Gd)
print("Ángulo entre G y d en grados:", (angulo_Gd * 180 / pi).evalf())

#3) Angulo entre H y d

norm_H=H.norm()

angulo_Hd=sp.acos(H.dot(d)/(norm_H*norm_d))
print("Angulo entre H y d:", angulo_Hd)
print("Ángulo entre H y d en grados:", (angulo_Hd * 180 / pi).evalf())
```
