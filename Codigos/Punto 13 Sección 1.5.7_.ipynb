```python
import sympy as sp

# Variables
t = sp.symbols('t', real=True)
x, y = sp.symbols('x y', real=True)

# Campo de fuerza
Fx = -y/(x**2 + y**2)
Fy = x/(x**2 + y**2)

# Parametrización del círculo de radio 1
x_t = sp.cos(t)
y_t = sp.sin(t)

# Derivadas respecto a t
dx_dt = sp.diff(x_t, t)
dy_dt = sp.diff(y_t, t)

# Sustitución en el campo
Fx_t = Fx.subs({x: x_t, y: y_t})
Fy_t = Fy.subs({x: x_t, y: y_t})

# Producto punto F·dr
F_dot_dr = Fx_t*dx_dt + Fy_t*dy_dt

# (a) De 0 a pi CCW (sentido antihorario)
trabajo_a = sp.integrate(F_dot_dr, (t, 0, sp.pi))

# (b) De 0 a -pi CW (sentido horario)
trabajo_b = sp.integrate(F_dot_dr, (t, 0, -sp.pi))

trabajo_a, trabajo_b
```
